```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import dask.dataframe as dd
#데이터가 너무 클 경우 사용하는 함수, pandas 대신에 사용한다. 좀 더 빠른 결과값을 도출해낼 수 있다.

%matplotlib inline  
#그래프를 주피터 화면에 띄워주는 기능을 함

import seaborn as sns
pd.options.display.max_columns = 999
pd.options.display.max_rows = 300
```

```python
df=pd.read_csv('CALL_CHICKEN_02MONTH.csv')
```

```python
df.head()
```

```python
seodaemun=df[(df['시군구']=='서대문구')]
```

```python
seodaemun
```

```python
seodaemun.info()
```

```python
seodaemun['기준일'] = pd.to_datetime(seodaemun['기준일'],format='%Y%m%d')
```

```python
p=seodaemun.set_index('기준일', inplace=False)
```

```python
p.info()
```

```python
p.to_csv('서대문구 읍면동 기준 통화건수.csv')
```

```python
plt.figure(figsize=[30,10],facecolor="white")
plt.plot(p['통화건수'].resample('D').sum(),'-',p['통화건수'].resample('D').sum(), "or")
plt.xlabel('Year_Month_Day',fontsize=12, rotation=0)
plt.title("Total number of call in Feb",fontsize=20, pad=35)
plt.grid(True,linewidth=3,linestyle='--',color='black',alpha=0.1)
```

```python
plt.figure(figsize=[50,10],facecolor='white')
p['통화건수'].resample('D').sum().plot(kind='bar')
plt.xlabel('Year_Month_Day',fontsize=12, rotation=0)
plt.title("Total number of call in Feb",fontsize=20, pad=35)
```

```python
d=p['통화건수'].resample('D').mean()
```

```python
d.to_csv('서대문구 일자별 총 전화건수.csv')
```
